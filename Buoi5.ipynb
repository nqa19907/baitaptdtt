```python
def Max_Of_Two( a, b):
  return max(a,b)
a,b = map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
def swap(a,b):
  return b,a
a,b = map(int,input().split())
print(*swap(a,b))
```

```python
def prime(n):
  for i in range(2,int(n**0.5)+1):
    if n%i==0:
      return False
  return True
print(prime(n=int(input())))
```

```python
n = int(input())
def perfect_Number(n):
  sum = 0
  for i in range(1,n//2+1):
    if n%i==0:
      sum += i
  if sum == n:
    return True
  return False
print(perfect_Number(n))
```

```python
def posi_Of_K(a,k):
  for i in range(len(a)):
    if int(a[i])==k:
      return i+1
  return -1
print(posi_Of_K(a=[i for i in input().split()],k = int(input())))
```

```python
def factorial(n):
  product = 1
  for i in range(1,n+1):
    product *= i
  return product
factorial(n = int(input()))
```

```python
a,b,c=map(str,input().split())
def calc(a,b,c):
  if b == '+':
    return round(float(a)+float(c),2)
  elif b == '-':
    return round(float(a)-float(c),2)
  elif b == '*':
    return round(float(a)*float(c),2)
  elif b == '/':
    return round(float(a)/float(c),2)
print(calc(a,b,c))
```

```python
a,b=map(int,input().split())
def hamming(a,b):
    m = bin(a^b)[2:]
    sum = 0
    for i in m:
        sum += int(i)
    return(sum)
print(hamming(a,b))
```

```python
n = input()
def sum_n(n):
  sum = 0
  for i in n:
    sum += int(i)
  return sum
print(sum_n(n))
```
