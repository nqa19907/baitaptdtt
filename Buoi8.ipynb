```python
#W4A5
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def init_empty(self):
        self.cart = []

    def add_item(self):
        name = input("Tên sản phẩm: ").strip()
        price = float(input("Giá: "))
        self.cart.append({"name": name, "price": price})

    def remove_item_by_name(self):
        name = input("Nhập tên sản phẩm cần xoá: ").strip()
        for i, item in enumerate(self.cart):
            if item["name"] == name:
                self.cart.pop(i)
                print("Đã xoá.")
                return
        print("Không tìm thấy sản phẩm.")

    def is_empty(self):
        return len(self.cart) == 0

    def total_price(self):
        return sum(item["price"] for item in self.cart)

    def show_cart(self):
        if self.is_empty():
            print("Giỏ hàng rỗng.")
            return
        for item in self.cart:
            print(f'- {item["name"]}: {item["price"]:.2f}')

    def clear_cart(self):
        self.cart.clear()


# ===== Main menu =====
cart = ShoppingCart()
cart.init_empty()

while True:
    print("""
2. Thêm sản phẩm
3. Xoá sản phẩm (theo tên)
4. Kiểm tra giỏ có rỗng không
5. Tính tổng tiền
6. Hiển thị giỏ hàng
7. Xoá toàn bộ giỏ hàng
""")

    choice = input("Chọn (2-7): ").strip()

    if choice == "2":
        cart.add_item()
    elif choice == "3":
        cart.remove_item_by_name()
    elif choice == "4":
        print("Rỗng" if cart.is_empty() else "Không rỗng")
    elif choice == "5":
        print(f"Tổng tiền: {cart.total_price():.2f}")
    elif choice == "6":
        cart.show_cart()
    elif choice == "7":
        cart.clear_cart()
        print("Đã xoá toàn bộ giỏ hàng.")
    else:
        print("Lựa chọn không hợp lệ.")

    out = input('Có muốn thoát khỏi chương trình không? (yes/no): ').strip().lower()
    if out == "yes":
        break
```

```python
#W4A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k
import sys

for line in sys.stdin:
    line = line.strip()
    if not line:
        continue
    w, h, k = map(float, line.split())

    r = Rectangle(w, h)
    r.scale(k)

    print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
#W4A7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Denominator cannot be zero")

        if den < 0:
            num = -num
            den = -den

        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g

    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)

    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)

    def __str__(self):
        return f"{self.num}/{self.den}"



a, b, op, c, d = input().split()

a = int(a)
b = int(b)
c = int(c)
d = int(d)

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    res = f1.add(f2)
elif op == "-":
    res = f1.sub(f2)
elif op == "*":
    res = f1.mul(f2)
elif op == "/":
    res = f1.div(f2)

print(res)

```

```python
#W4A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if self.balance >= x:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"



owner, initial_balance = input().split()
account = BankAccount(owner, float(initial_balance))

n = int(input())

for _ in range(n):
    cmd, value = input().split()
    value = float(value)

    if cmd == "DEPOSIT":
        account.deposit(value)
    elif cmd == "WITHDRAW":
        account.withdraw(value)

print(f"{account.balance:.2f}")
```

```python
#W4A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0.0

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        if a >= 6.5:
            return "Good"
        if a >= 5:
            return "Average"
        return "Poor"


name = input().strip()
n = int(input().strip())
st = Student(name)

for _ in range(n):
    subject, score = input().split()
    st.add_score(subject, score)

a = st.avg()
print(f"{st.name} {a:.2f} {st.rank()}")
```

```python
#W4A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)


lib = Library()
n = int(input())

for _ in range(n):
    line = input().strip()
    if line.startswith("ADD"):
        _, rest = line.split(" ", 1)
        title, author, year = rest.split(";")
        lib.add(Book(title, author, year))
    elif line.startswith("COUNTYEAR"):
        _, y = line.split()
        print(lib.find_by_year(int(y)))
    elif line.startswith("COUNT"):
        _, author = line.split()
        print(lib.count_by_author(author))
```

```python
#W4A11
import math

class P:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z

    @staticmethod
    def sub(a, b):
        return (a.x - b.x, a.y - b.y, a.z - b.z)

class A:
    @staticmethod
    def cr(u, v):
        return (u[1]*v[2]-u[2]*v[1], u[2]*v[0]-u[0]*v[2], u[0]*v[1]-u[1]*v[0])

    @staticmethod
    def dt(u, v):
        return u[0]*v[0] + u[1]*v[1] + u[2]*v[2]

    @staticmethod
    def nm(u):
        return math.sqrt(A.dt(u, u))

    @staticmethod
    def ang(a, b, c, d):
        ba = P.sub(b, a)
        ca = P.sub(c, a)
        cb = P.sub(c, b)
        db = P.sub(d, b)
        n1 = A.cr(ba, ca)
        n2 = A.cr(cb, db)
        den = A.nm(n1) * A.nm(n2)
        if den == 0:
            return 0.0
        x = abs(A.dt(n1, n2)) / den
        x = max(-1.0, min(1.0, x))
        return math.degrees(math.acos(x))

a = P(*map(float, input().split()))
b = P(*map(float, input().split()))
c = P(*map(float, input().split()))
d = P(*map(float, input().split()))
print(f"{A.ang(a,b,c,d):.2f}")
```

```python
#W4A13
import math

class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = int(a), int(b), int(c)

    def area(self):
        a, b, c = self.a, self.b, self.c
        if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
            return -1
        s = (a + b + c) / 2
        return math.sqrt(s * (s - a) * (s - b) * (s - c))

a, b, c = map(int, input().split())
t = Triangle(a, b, c)
x = t.area()
print("invalid" if x < 0 else f"{x:.2f}")
```
