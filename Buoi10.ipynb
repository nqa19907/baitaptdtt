```python
#1
def merge(a,b):
    i=j=0
    res=[]
    while i<len(a) and j < len(b):
        if a[i]>=b[j]:
            res+=[b[j]]
            j+=1
        elif a[i]<=b[j]:
            res+=[a[i]]
            i+=1
    res+=a[i:]
    res+=b[j:]
    return res
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

```python
#2
def kth_largest(a, k):
    n = len(a)
    m = len(a[0])
    lo = a[0][0]
    hi = a[n-1][m-1]
    while lo < hi:
        mid = (lo + hi + 1) // 2
        cnt = 0
        j = 0
        for i in range(n-1, -1, -1):
            while j < m and a[i][j] < mid:
                j += 1
            cnt += m - j
        if cnt >= k:
            lo = mid
        else:
            hi = mid - 1
    return lo
n, m, k = map(int, input().split())
a = []
for i in range(n):
    a.append(list(map(int, input().split())))
print(kth_largest(a, k))
```

```python
#3
def lst_to_str(a):
    m=''
    for i in a:
        m+=str(i)
    return m
def str_to_lst(a):
    m=[]
    for i in a:
        m+=[int(i)]
    return m
a=input()
minn=min(str_to_lst(a))
arr=str_to_lst(a)
swap=False
for i in range(len(arr)-1):
    for j in range(i+1,len(arr)):
        if arr[j]==minn:
            arr[j],arr[i]=arr[i],arr[j]
            swap=True
            break
    if swap: break
if !swap:
    arr[-1],arr[-2]=arr[-2],arr[-1]
print(int(lst_to_str(arr)))
```

```python
#4
def can_transform(a, b):
    n = len(a)
    m = len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        if a[i-1].islower():
            dp[i][0] = dp[i-1][0]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if a[i-1].upper() == b[j-1]:
                dp[i][j] = dp[i-1][j-1]
            if a[i-1].islower():
                dp[i][j] = dp[i][j] or dp[i-1][j]
    return dp[n][m]

q = int(input())
for _ in range(q):
    a = input().strip()
    b = input().strip()
    print("YES" if can_transform(a, b) else "NO")
```

```python
#7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    i = 0
    while i < n and k > 0:
        if flowerbed[i] == 0:
            left = 0 if i == 0 else flowerbed[i - 1]
            right = 0 if i == n - 1 else flowerbed[i + 1]
            if left == 0 and right == 0:
                flowerbed[i] = 1
                k -= 1
                i += 2
                continue
        i += 1
    return k == 0
```

```python
#8
def is_path_crossing(moves):
    x = y = 0
    seen = {(0, 0)}
    for c in moves:
        if c == 'R':
            x += 1
        elif c == 'L':
            x -= 1
        elif c == 'U':
            y += 1
        elif c == 'D':
            y -= 1
        if (x, y) in seen:
            return True
        seen.add((x, y))
    return False
```

```python
#9
def Median(nums):
    a, b, c, d, e = nums
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > e: b, e = e, b
    if b > c: b, c = c, b
    if b > d: b, d = d, b
    return b
```

```python
#10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    i = 0
    res = 0
    for h in houses:
        while i + 1 < len(heaters) and abs(heaters[i + 1] - h) <= abs(heaters[i] - h):
            i += 1
        d = abs(heaters[i] - h)
        if d > res:
            res = d
    return res
```
