```python
n = input()
sum = 0
palin = int(n)
def isPalindrome(k):
    palindrome = True
    for i in range(len(k)):
        if k[i] != k[len(k)-1-i]:
            palindrome = False
    return palindrome
def reverse(n):
    rev = ''
    for i in range(len(n)-1,-1,-1):
        rev = rev + n[i]
    return rev
while isPalindrome(str(palin)) == False:
    sum += 1
    palin = palin + int(reverse(str(palin)))
print(sum,palin)
```

```python
def pw(n):
    if len(n)<=6:
        return False
    caps = 0
    nor = 0
    sym = 0
    num = 0
    for i in n:
        if 'a'<=i<='z':
            nor += 1
        elif 'A'<=i<='Z':
            caps += 1
        elif '0'<=i<='9':
            num += 1
        elif 33<=ord(i)<=47 or 58<=ord(i)<=64 or 91<=ord(i)<=96 or 123<=ord(i)<=126:
            sym += 1
    if sym<1 or nor <1 or caps<1 or num <1:
        return False
    return True
print(pw(n=input()))
```

```python
n = int(input())
n = str(n)
m=''
num =''
def add_Dot(n):
    m=''
    num=''
    for i in range(2,len(n),3):
        for j in range(i-2,i+1):
            m += n[j]
        num +=m
        if i !=len(n)-1:
            num +='.'
        m=''
    return num
if len(n)>3:
    if len(n)%3==0:
        print(add_Dot(n))
    elif (len(n)-2)%3==0:
        for i in range(0,2):
            num += n[i]
        num+='.'
        for i in range(2,len(n)):
            m += n[i]
        num += add_Dot(m)
        print(num)
    else:
        for i in range(0,1):
            num += n[i]
        num+='.'
        for i in range(1,len(n)):
            m += n[i]
        num += add_Dot(m)
        print(num)
else: print(n)
```
