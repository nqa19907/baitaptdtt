```python
#12
n=list(map(int,input().split()))
def cc(n):
    a={}
    for i in n:
        if i not in a:
            a[i]=1
        else:
            a[i]+=1
    max=0
    b=[]
    for i in a:
        if a[i]>=max:
            max=a[i]
            idx=i
            b.append(i)
    for i in a:
        for j in a:
            if a[i]!=a[j]:
                return idx
    return min(b)
print(cc(n))
```
