```python
#1
try:
    while True:
        a, b = map(int, input().split())
        try:
            print(f"{a/b:.2f}")
        except ZeroDivisionError:
            print("Loi chia khong")
except:
    pass
```

```python
#2
try:
    n = int(input())
    my_list = list(map(int, input().split()))
    try:
        print(my_list[n])
    except IndexError:
        print("IndexError")
except:
    pass
```

```python
#3
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum((int(c) ** 2 for c in str(n)))
    return n == 1

while True:
    try:
        a = int(input())
        print("YES" if is_happy(a) else "NO")
    except ValueError:
        break
```

```python
#6
import math

while True:
    try:
        x = float(input())
        if x < 0:
            raise ValueError
        print(f"{math.sqrt(x):.2f}")
    except ValueError:
        print("So am")
    except:
        break
```
